```python
import pandas as pd

df_clustering_file = pd.read_csv("../outputs/sofia_product_clusters.csv")
# print(df_clustering_file.head())
print(df_clustering_file.shape)
df_clustering_file.head()
```

```python
df_classification_file = pd.read_csv("../outputs/data_with_sentiment_clean.csv")
# print(df_clustering_file.head())
print(df_classification_file.shape)
df_classification_file.head()

df_classification_cleaned = df_classification_file[["id", "sentiment"]]
df_classification_cleaned.head()
```

```python
import pandas as pd

# Remove duplicate IDs from classification file
df_classification_unique = df_classification_cleaned.drop_duplicates(subset=["id"])

# Merge (like enriching clustering file with sentiment info)
df_merged = pd.merge(
    df_clustering_file,
    df_classification_unique,
    left_on="ProductID",
    right_on="id",
    how="left"  # keep all clustering rows
)

# Drop duplicate key column (optional)
df_merged = df_merged.drop(columns=["id"])

# Check shape
print(df_merged.shape)
print(df_merged.head())
```
